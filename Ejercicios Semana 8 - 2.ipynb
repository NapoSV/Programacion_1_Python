# Parte 1 — Conceptos Básicos

Antes de resolver los ejercicios es importante entender qué es cada estructura:

| Estructura | Descripción | Característica clave |
|---|---|---|
| Tupla | Colección ordenada e inmutable | No se puede modificar |
| Diccionario | Colección de pares clave‑valor | Acceso rápido por clave |
| Set | Conjunto de elementos únicos | No permite duplicados |
| Array | Lista optimizada de elementos del mismo tipo | Más eficiente para datos numéricos |

````python
# 16 Promedio estudiantes

estudiantes = {
"Ana":90,
"Luis":80,
"Pedro":85
}

promedio = sum(estudiantes.values())/len(estudiantes)

print(promedio)
````

## Sets

````python
# 19 Unión intersección diferencia

A = {1,2,3,4}
B = {3,4,5,6}

print("Union:",A|B)
print("Interseccion:",A&B)
print("Diferencia:",A-B)
````

````python
# 20 Eliminar duplicados

lista = [1,2,2,3,4,4,5]

sin_duplicados = list(set(lista))

print(sin_duplicados)
````

````python
# 21 Subconjunto

print({1,2}.issubset({1,2,3,4,5}))
````

## Arrays

````python
# 22 Multiplicar elementos

from array import array

arr = array('i',[2,4,6,8,10])

resultado = array('i',[x*2 for x in arr])

print(resultado)
````

````python
# 23 Mayor y menor

def mayor_menor(arr):
    mayor = arr[0]
    menor = arr[0]

    for n in arr:
        if n > mayor:
            mayor = n
        if n < menor:
            menor = n

    return mayor, menor

print(mayor_menor([3,8,1,9,2]))
````

````python
# 24 Invertir array manualmente

arr = [1,2,3,4,5]

invertido = []

for i in range(len(arr)-1,-1,-1):
    invertido.append(arr[i])

print(invertido)
````

# Ejercicios Avanzados

## Diccionarios

````python
# 28 Buscar producto

productos = {
"pan":0.5,
"leche":1.2,
"huevos":2.5
}

producto = "pan"

print(productos.get(producto,"No existe"))
````

````python
# 29 Frecuencia de palabras

texto = "python es facil y python es poderoso"

frecuencia = {}

for palabra in texto.split():
    frecuencia[palabra] = frecuencia.get(palabra,0)+1

print(frecuencia)
````

````python
# 30 Agenda telefonica

agenda = {}

agenda["Juan"] = ["1234-5678","8888-9999"]
agenda["Maria"] = ["7777-1111"]

print(agenda)
````

## Sets

````python
# 31 Anagramas

def son_anagramas(p1,p2):
    return set(p1) == set(p2)

print(son_anagramas("roma","amor"))
````

````python
# 32 Letras únicas entre frases

f1 = "hola mundo"
f2 = "python es genial"

unicas = set(f1) ^ set(f2)

print(unicas)
````

````python
# 33 Pares al cuadrado

nums = {1,2,3,4,5,6}

resultado = {x**2 for x in nums if x%2==0}

print(resultado)
````

## Arrays

````python
# 34 Media mediana moda

import statistics

datos = [1,2,2,3,4]

print("Media:",statistics.mean(datos))
print("Mediana:",statistics.median(datos))
print("Moda:",statistics.mode(datos))
````

````python
# 35 Dos numeros mas grandes

arr = [3,9,5,8,7]

mayor1 = mayor2 = float('-inf')

for n in arr:
    if n > mayor1:
        mayor2 = mayor1
        mayor1 = n
    elif n > mayor2:
        mayor2 = n

print(mayor1,mayor2)
````

````python
# 36 Matriz 3x3 con array

from array import array

matriz = [
array('i',[1,2,3]),
array('i',[4,5,6]),
array('i',[7,8,9])
]

for fila in matriz:
    print("Suma fila:",sum(fila))
````

# Tips Importantes para Programar en Python

### 1. Usa `len()` para conocer el tamaño de estructuras
Esto es fundamental cuando trabajas con listas, tuplas o sets.

### 2. Usa `.get()` en diccionarios
Evita errores cuando una clave no existe.

```python
dic.get("clave", "valor por defecto")
```

### 3. Sets son ideales para:
- eliminar duplicados
- operaciones matemáticas de conjuntos
- comparaciones rápidas

### 4. Las tuplas son útiles cuando:
- los datos **no deben cambiar**
- se quiere mejorar rendimiento

### 5. Comprensiones de Python
Son una forma muy poderosa de escribir código más limpio:

```python
[x*2 for x in lista]
```

### 6. Buenas prácticas
- Usa nombres de variables claros
- Comenta el código
- Divide problemas en funciones

### 7. Consejo profesional

Si quieres trabajar en **Data Science o BI (como Power BI o análisis de datos)**, dominar estas estructuras es fundamental porque:

- los **diccionarios** representan JSON y APIs
- los **sets** ayudan en análisis de datos
- las **tuplas** se usan mucho en consultas SQL
- los **arrays/listas** son la base de pandas y numpy
