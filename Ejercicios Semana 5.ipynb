11. Pedir al usuario un número y mostrar sus divisores usando un for
• Descripción: Usa un for desde 1 hasta el número ingresado.
• Instrucciones:
o Si num % i == 0, entonces i es divisor.
• Punto clave: Evita divisiones por 0 y asegúrate de que el número sea positivo.

```python
numeroUsuario = int(input("Ingrese un número mayor a cero"))

while numeroUsuario <= 0:
    numeroUsuario = int(input("Ingrese un número mayor a cero"))

for numero in range(1,numeroUsuario):
    if numeroUsuario % numero == 0:
        print(f'El número {numero} es divisor de {numeroUsuario}')
```

12. Determinar si un número ingresado por el usuario es primo usando un while
• Descripción: Un número primo solo tiene dos divisores (1 y él mismo).
• Instrucciones:
o Usa un while para verificar cuántos divisores tiene el número.
• Punto clave: Si encuentra más de dos divisores, ya no es primo.

```python
numeroUsuario = int(input("Ingrese un número mayor a cero:"))

while numeroUsuario <= 0:
    numeroUsuario = int(input("Ingrese un número mayor a cero:"))

divisoresEncontrados = 0
divisorActual = 1

while divisorActual <= numeroUsuario:
    if numeroUsuario % divisorActual == 0:
        divisoresEncontrados = divisoresEncontrados + 1
        if divisoresEncontrados > 2:
            break
    divisorActual = divisorActual + 1

if divisoresEncontrados == 2:
    print(f"{numeroUsuario} es primo")
else:
    print(f"{numeroUsuario} no es primo")

print(divisoresEncontrados)
```

15. Calcular el factorial de un número ingresado por el usuario
• Descripción: Usa un for desde 1 hasta el número ingresado.
• Instrucciones:
o Multiplica en cada iteración para obtener el factorial.
• Punto clave: Factorial de 0 es 1 por definición.

```python
numeroUsuario = int(input("Ingrese un número para calcular su factorial:"))

factorial = 1
contador = 1
if numeroUsuario == 1 or numeroUsuario == 0:
    print(f'El factorial de {numeroUsuario} por definición es: 1')
else:
    for numero in range(1,numeroUsuario + 1):
        factorial = factorial * numero

print(f'El factorial (n!) de {numeroUsuario} es: {factorial}')
```

16. Imprimir la secuencia de Fibonacci hasta un número ingresado por el usuario
• Descripción: La secuencia de Fibonacci empieza en 0, 1, 1, 2, 3, 5, ...
• Instrucciones:
o Usa dos variables para almacenar los dos últimos números de la secuencia.
o Continúa sumando hasta que el número sea mayor al ingresado.
• Punto clave: Usa un while para continuar mientras no se pase del límite.
Nota: El siguiente número es la suma de los dos anteriores

```python
limiteUsuario = int(input("Ingrese el límite para la secuencia Fibonacci:"))

numeroA = 0
numeroB = 1

print(numeroA)

while limiteUsuario < 0:
    limiteUsuario = int(input("Ingrese un número positivo Fibonacci:"))
    
while numeroB <= limiteUsuario:
    print(numeroB)
    secuencia = numeroA + numeroB
    numeroA = numeroB
    numeroB = secuencia
    
```
