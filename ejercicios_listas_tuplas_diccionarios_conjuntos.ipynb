# 🔷 SECCIÓN 1: Listas (`list`)

Las **listas** son una de las estructuras de datos más utilizadas en Python. Se caracterizan por:

- ✅ Son **ordenadas**: los elementos mantienen su posición.
- ✅ Son **mutables**: se pueden modificar después de crearse.
- ✅ Permiten **elementos duplicados**.
- ✅ Pueden contener **tipos de datos mixtos**.

Se definen con corchetes `[]` y los elementos se separan por comas.

```python
mi_lista = [10, 20, 30, 40, 50]
```

### 📌 Índices en Python
Los índices comienzan en `0`. También se pueden usar **índices negativos** para acceder desde el final:

| Elemento | 10 | 20 | 30 | 40 | 50 |
|----------|----|----|----|----|----|
| Índice + | 0  | 1  | 2  | 3  | 4  |
| Índice - | -5 | -4 | -3 | -2 | -1 |

## ✏️ Ejercicio 1: Acceso y Modificación de Listas

### Conceptos clave:

- **Acceso por índice**: `lista[0]` devuelve el primer elemento, `lista[-1]` devuelve el último.
- **Modificación**: como las listas son mutables, se puede reasignar un valor usando su índice: `lista[1] = nuevo_valor`.

### ¿Por qué `lista[-1]` para el último elemento?
Usar `lista[-1]` es preferible a `lista[len(lista)-1]` porque es más limpio y no requiere conocer el tamaño de la lista. Es una práctica muy común en Python (**Pythonic**).

```python
# Acceso
lista[0]   # Primer elemento
lista[-1]  # Último elemento

# Modificación
lista[1] = 99  # Cambia el segundo elemento a 99
```

````python
"""
Acceso y Modificación: Crea una lista de 5 números enteros.
Imprime el primer y el último número de la lista.
Luego, cambia el segundo número a otro valor y muestra la lista modificada.
"""

# Creación de lista
numeros = [6, 1, 60, 55, 31]
print("Lista original:", numeros)

# Imprime el primer y el último número de la lista
print("Primer número:", numeros[0])
print("Último número:", numeros[-1])

# cambia el segundo número a otro valor y muestra la lista modificada (indice 1)
numeros[1] = 99
print("Lista modificada (segundo número cambiado a 99):", numeros)
````

## ✏️ Ejercicio 2: Operaciones con Listas

### Conceptos clave:

Python ofrece métodos integrados para manipular listas:

| Método | Descripción | Ejemplo |
|--------|-------------|--------|
| `.append(x)` | Agrega `x` al **final** de la lista | `lista.append("mango")` |
| `.insert(i, x)` | Inserta `x` en la posición `i` | `lista.insert(1, "uva")` |
| `.remove(x)` | Elimina la **primera** ocurrencia de `x` | `lista.remove("banana")` |
| `.pop(i)` | Elimina y retorna el elemento en posición `i` | `lista.pop(0)` |
| `.sort()` | Ordena la lista in-place | `lista.sort()` |

> ⚠️ `.remove(x)` lanza un `ValueError` si el elemento no existe en la lista.

````python
# Ejercicio 2: Operaciones con Listas

frutas = ["manzana", "banana", "cereza", "kiwi"]
print("Lista original:", frutas)

# Agregamos "mango" al final de la lista
frutas.append("mango")
print("\nDespués de agregar 'mango':", frutas)

# Eliminamos "banana" de la lista
frutas.remove("banana")
print("Después de eliminar 'banana':", frutas)
````

## ✏️ Ejercicio 3: Slicing (Rebanado) de Listas

### Conceptos clave:

El **slicing** permite extraer porciones de una lista usando la sintaxis:

```python
lista[inicio:fin:paso]
```

- `inicio`: índice donde comienza el slice (incluido).
- `fin`: índice donde termina el slice (**excluido** — no se incluye).
- `paso`: de cuánto en cuánto se toman los elementos (opcional, por defecto 1).

### Ejemplos visuales:
```python
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
#índices:   0  1  2  3  4  5  6  7  8

numeros[2:5]   # → [3, 4, 5]  (índices 2, 3, 4)
numeros[::2]   # → [1, 3, 5, 7, 9]  (cada dos elementos)
numeros[::-1]  # → [9, 8, 7, 6, 5, 4, 3, 2, 1]  (invertida)
```

> 💡 El índice `fin` es **exclusivo**, lo que significa que `lista[2:5]` toma los índices 2, 3 y 4, pero **no** el 5.

````python
# Ejercicio 3: Slicing

numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print("Lista completa:", numeros)

# Slicing del índice 2 al 5 (sin incluir el índice 5)
# Esto nos dará los elementos en las posiciones 2, 3 y 4
resultado = numeros[2:5]
print("\nSlice [2:5] (índices 2, 3 y 4):", resultado)

# Bonus: otros ejemplos de slicing
print("\n--- Ejemplos adicionales de slicing ---")
print("Primeros 4 elementos [0:4]:", numeros[0:4])
print("Últimos 3 elementos [-3:]:", numeros[-3:])
print("Lista invertida [::-1]:", numeros[::-1])
print("Cada 2 elementos [::2]:", numeros[::2])
````

## ✏️ Ejercicio 4: Comprensión de Listas (List Comprehension)

### Conceptos clave:

La **comprensión de listas** es una forma concisa y elegante de crear listas en Python. Su sintaxis es:

```python
nueva_lista = [expresión for elemento in iterable if condición]
```

- `expresión`: lo que se agrega a la lista (puede ser una operación).
- `elemento`: variable que toma cada valor del iterable.
- `iterable`: la secuencia sobre la que se itera.
- `if condición`: (opcional) filtro para incluir solo ciertos elementos.

### Comparación con el bucle tradicional:

```python
# Forma tradicional (bucle for)
cuadrados = []
for n in range(1, 11):
    cuadrados.append(n ** 2)

# Forma Pythonic (comprensión de listas) ✅
cuadrados = [n ** 2 for n in range(1, 11)]
```

> 💡 La comprensión de listas es generalmente **más rápida** y **más legible** que un bucle `for` equivalente. Es una de las características más valoradas de Python.

````python
# Ejercicio 4: Comprensión de Listas

# Creamos la lista de cuadrados del 1 al 10 usando list comprehension
# n**2 eleva n a la potencia 2 (n al cuadrado)
cuadrados = [n ** 2 for n in range(1, 11)]
print("Cuadrados del 1 al 10:", cuadrados)

# Mostramos de forma más detallada
print("\n--- Detalle de cada cuadrado ---")
for n in range(1, 11):
    print(f"{n}² = {n**2}")

# Bonus: solo cuadrados de números pares usando condición
cuadrados_pares = [n ** 2 for n in range(1, 11) if n % 2 == 0]
print("\nCuadrados de números pares (1-10):", cuadrados_pares)
````

# 🔶 SECCIÓN 2: Tuplas (`tuple`)

Las **tuplas** son similares a las listas, pero con una diferencia fundamental: son **inmutables**.

| Característica | Lista `[]` | Tupla `()` |
|---------------|-----------|----------|
| Ordenada | ✅ | ✅ |
| Mutable | ✅ | ❌ |
| Duplicados | ✅ | ✅ |
| Velocidad | Normal | Más rápida |
| Uso típico | Datos cambiantes | Datos fijos/constantes |

Se definen con paréntesis `()` o simplemente con comas:

```python
mi_tupla = (1, "hola", 3.14)
# También válido:
mi_tupla = 1, "hola", 3.14
```

### ¿Cuándo usar tuplas?
- Para almacenar datos que **no deben cambiar** (coordenadas, colores RGB, etc.).
- Como **claves de diccionarios** (las listas no pueden ser claves).
- Cuando se necesita mayor **rendimiento** que una lista.

## ✏️ Ejercicio 5: Tupla con Diferentes Tipos de Datos

### Conceptos clave:

Una tupla puede contener **cualquier tipo de dato** en Python: enteros (`int`), cadenas de texto (`str`), flotantes (`float`), booleanos (`bool`), incluso otras listas o tuplas.

El acceso a los elementos es igual que en las listas, usando **índices**:

```python
mi_tupla = (42, "Python", 3.14)
#índices:    0      1       2

mi_tupla[0]  # → 42      (entero)
mi_tupla[1]  # → "Python" (cadena)
mi_tupla[2]  # → 3.14    (flotante)
```

> ⚠️ Si intentas modificar un elemento de una tupla (`mi_tupla[0] = 99`), Python lanzará un `TypeError: 'tuple' object does not support item assignment`.

````python
# Ejercicio 5: Tupla con Diferentes Tipos de Datos

# Creamos una tupla con 3 tipos diferentes: int, str, float
mi_tupla = (42, "Python", 3.14)
print("Tupla completa:", mi_tupla)

# Verificamos el tipo de cada elemento
print("\n--- Tipos de datos en la tupla ---")
for i, elemento in enumerate(mi_tupla):
    print(f"  Índice {i}: {elemento!r} → tipo: {type(elemento).__name__}")

# Accedemos al segundo elemento (índice 1)
print("\nSegundo elemento (índice 1):", mi_tupla[1])

# Demostramos la inmutabilidad
print("\n--- Demostración de inmutabilidad ---")
try:
    mi_tupla[0] = 99  # Esto generará un error
except TypeError as e:
    print(f"Error al intentar modificar la tupla: {e}")
````

## ✏️ Ejercicio 6: Acceso y Slicing en Tuplas

### Conceptos clave:

Aunque las tuplas son **inmutables** (no se pueden modificar), sí permiten:
- **Acceso por índice**: igual que en listas.
- **Slicing**: extraer subconjuntos de elementos.

El slicing en tuplas devuelve **otra tupla**, no una lista:

```python
colores = ("rojo", "verde", "azul", "amarillo")
#índices:    0        1       2        3

colores[0]    # → "rojo"              (primer elemento)
colores[-1]   # → "amarillo"          (último elemento)
colores[0:3]  # → ("rojo", "verde", "azul")  (primeros 3)
```

### Funciones útiles con tuplas:
- `len(tupla)` → número de elementos
- `tupla.count(x)` → cuántas veces aparece `x`
- `tupla.index(x)` → índice de la primera ocurrencia de `x`

````python
# Ejercicio 6: Acceso y Slicing en Tuplas

colores = ("rojo", "verde", "azul", "amarillo")
print("Tupla de colores:", colores)
print("Número de elementos:", len(colores))

# Acceso al primer elemento (índice 0) y al último (índice -1)
print("\nPrimer elemento (índice 0):", colores[0])
print("Último elemento (índice -1):", colores[-1])

# Slicing: primeros tres elementos [0:3]
primeros_tres = colores[0:3]
print("\nPrimeros tres elementos [0:3]:", primeros_tres)
print("Tipo del resultado del slicing:", type(primeros_tres))

# Bonus: más operaciones con la tupla
print("\n--- Operaciones adicionales ---")
print("Índice de 'azul':", colores.index("azul"))
print("Tupla invertida:", colores[::-1])
````

# 🟢 SECCIÓN 3: Diccionarios (`dict`)

Los **diccionarios** son colecciones de pares **clave: valor**. Son la estructura de datos más potente para almacenar información relacionada.

### Características:
- ✅ **No ordenados** (en Python 3.7+ mantienen el orden de inserción).
- ✅ **Mutables**: se pueden agregar, modificar y eliminar pares.
- ✅ Las **claves deben ser únicas** e inmutables (strings, números, tuplas).
- ✅ Los **valores** pueden ser cualquier tipo.

```python
# Creación
persona = {
    "nombre": "Ana",
    "edad": 30,
    "ciudad": "San Salvador"
}

# Acceso
persona["nombre"]       # → "Ana"
persona.get("edad")     # → 30 (más seguro, no lanza error)

# Modificación
persona["edad"] = 31    # Actualiza el valor

# Agregar nueva clave
persona["email"] = "ana@email.com"

# Eliminar
del persona["ciudad"]   # Elimina la clave y su valor
```

## ✏️ Ejercicio 7: Diccionario de Información Personal

### Conceptos clave:

Al trabajar con diccionarios, existen **dos formas** de acceder a un valor:

```python
# Forma 1: Acceso directo (puede lanzar KeyError si no existe)
diccionario["clave"]

# Forma 2: Método .get() (devuelve None si no existe, más seguro)
diccionario.get("clave")
diccionario.get("clave", "valor_por_defecto")
```

Para **modificar un valor** existente, simplemente se reasigna:

```python
info["edad"] = 26  # Si "edad" existe, la actualiza; si no, la crea
```

### Métodos útiles:
- `.keys()` → todas las claves
- `.values()` → todos los valores
- `.items()` → pares (clave, valor)

````python
# Ejercicio 7: Diccionario de Información

# Creamos el diccionario con las claves requeridas
info = {
    "nombre": "María García",
    "edad": 25,
    "ciudad": "San Salvador"
}
print("Diccionario original:")
for clave, valor in info.items():
    print(f"  {clave}: {valor}")

# Accedemos al valor de la clave "nombre"
print("\nValor de 'nombre':", info["nombre"])

# Cambiamos el valor de "edad"
info["edad"] = 26
print("\nDiccionario después de cambiar 'edad' a 26:")
for clave, valor in info.items():
    print(f"  {clave}: {valor}")

# Mostramos claves y valores por separado
print("\nClaves disponibles:", list(info.keys()))
print("Valores disponibles:", list(info.values()))
````

## ✏️ Ejercicio 8: Agregar y Eliminar Claves en un Diccionario

### Conceptos clave:

**Agregar** una clave nueva es igual que modificar una existente — Python detecta si la clave ya existe:

```python
persona["ciudad"] = "San Salvador"  # Agrega si no existe, actualiza si existe
```

**Eliminar** claves tiene tres opciones:

| Método | Descripción | ¿Error si no existe? |
|--------|-------------|--------------------|
| `del diccionario["clave"]` | Elimina la clave | ✅ Sí (`KeyError`) |
| `diccionario.pop("clave")` | Elimina y **retorna** el valor | ✅ Sí |
| `diccionario.pop("clave", None)` | Elimina sin error si no existe | ❌ No |

> 💡 Usa `.pop()` cuando necesites el valor eliminado para usarlo después.

````python
# Ejercicio 8: Agregar y Eliminar Claves

persona = {"nombre": "Carlos", "edad": 25}
print("Diccionario original:", persona)

# Agregamos la clave "ciudad" con el valor "San Salvador"
persona["ciudad"] = "San Salvador"
print("\nDespués de agregar 'ciudad':", persona)

# Eliminamos la clave "edad" usando del
del persona["edad"]
print("Después de eliminar 'edad':", persona)

# Bonus: demostración de .pop() que retorna el valor eliminado
persona2 = {"nombre": "Laura", "edad": 30, "ciudad": "Santa Ana"}
edad_eliminada = persona2.pop("edad")
print("\n--- Bonus: usando .pop() ---")
print(f"Valor eliminado con .pop(): {edad_eliminada}")
print("Diccionario resultante:", persona2)
````

# 🔴 SECCIÓN 4: Conjuntos (`set`)

Los **conjuntos** son colecciones **no ordenadas** de elementos **únicos**. Están inspirados en los conjuntos matemáticos.

### Características:
- ✅ **No permiten duplicados** (los eliminan automáticamente).
- ❌ **No están ordenados** (no tienen índices).
- ✅ Son **mutables** (se pueden agregar/eliminar elementos).
- ✅ Optimizados para operaciones de **búsqueda rápida**.

```python
mi_set = {1, 2, 3, 4, 5}
# También: set([1, 2, 3])  ← desde una lista
```

### Operaciones de conjuntos (como en matemáticas):

| Operación | Símbolo | Método |
|-----------|---------|--------|
| Unión | `A \| B` | `A.union(B)` |
| Intersección | `A & B` | `A.intersection(B)` |
| Diferencia | `A - B` | `A.difference(B)` |
| Diferencia simétrica | `A ^ B` | `A.symmetric_difference(B)` |

## ✏️ Ejercicio 9: Operaciones con Conjuntos

### Conceptos clave:

Para modificar un conjunto se usan estos métodos:

```python
mi_set.add(x)       # Agrega el elemento x (sin error si ya existe)
mi_set.remove(x)    # Elimina x (lanza KeyError si no existe)
mi_set.discard(x)   # Elimina x (NO lanza error si no existe ✅)
mi_set.pop()        # Elimina y retorna un elemento aleatorio
```

### ¿Por qué usar `discard` en lugar de `remove`?
- `remove(x)` lanza `KeyError` si `x` no está en el conjunto.
- `discard(x)` es más **seguro** porque no falla si el elemento no existe.

> ⚠️ Recuerda que los conjuntos **no tienen orden garantizado**, por lo que al imprimir el resultado, el orden puede variar.

````python
# Ejercicio 9: Operaciones con Conjuntos

# Creamos el conjunto inicial
numeros = {1, 2, 3, 4, 5}
print("Conjunto original:", numeros)

# Agregamos el número 6
numeros.add(6)
print("\nDespués de agregar 6:", numeros)

# Eliminamos el número 3
numeros.remove(3)
print("Después de eliminar 3:", numeros)

# Bonus: intentar agregar un duplicado
numeros.add(6)  # No genera error, simplemente no hace nada
print("\nDespués de intentar agregar 6 nuevamente (duplicado):", numeros)
print("(Los conjuntos ignoran automáticamente los duplicados)")

# Bonus: diferencia entre remove y discard
print("\n--- remove vs discard ---")
numeros.discard(99)  # No lanza error aunque 99 no existe
print("discard(99) ejecutado sin error (99 no estaba en el conjunto)")
print("Conjunto final:", numeros)
````

## ✏️ Ejercicio 10: Intersección de Conjuntos

### Conceptos clave:

La **intersección** de dos conjuntos A y B devuelve un nuevo conjunto con los elementos que están en **ambos** conjuntos simultáneamente:

```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

A ∩ B = {3, 4}  ← solo los elementos comunes
```

### Dos formas equivalentes:

```python
# Usando el operador &
comunes = A & B

# Usando el método .intersection()
comunes = A.intersection(B)
```

### Visualización de operaciones de conjunto:

```
A = {1, 2, 3, 4}    B = {3, 4, 5, 6}
    ┌─────────────────────┐
    │  1  2  │ 3  4 │  5  6  │
    │  solo A │ A∩B  │ solo B │
    └─────────────────────┘
```

| Operación | Resultado con A={1,2,3,4} y B={3,4,5,6} |
|-----------|------------------------------------------|
| `A & B` (intersección) | `{3, 4}` |
| `A \| B` (unión) | `{1, 2, 3, 4, 5, 6}` |
| `A - B` (diferencia) | `{1, 2}` |
| `B - A` (diferencia) | `{5, 6}` |
| `A ^ B` (dif. simétrica) | `{1, 2, 5, 6}` |

````python
# Ejercicio 10: Intersección de Conjuntos

# Creamos los dos conjuntos
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}

print("Conjunto A:", conjunto_a)
print("Conjunto B:", conjunto_b)

# Encontramos la intersección (elementos comunes)
interseccion = conjunto_a.intersection(conjunto_b)
print("\nIntersección (A ∩ B):", interseccion)

# Forma alternativa usando el operador &
interseccion_op = conjunto_a & conjunto_b
print("Intersección con operador &:", interseccion_op)

# Bonus: todas las operaciones de conjunto
print("\n--- Todas las operaciones de conjunto ---")
print(f"A = {conjunto_a}")
print(f"B = {conjunto_b}")
print(f"Unión (A | B):                {conjunto_a | conjunto_b}")
print(f"Intersección (A & B):         {conjunto_a & conjunto_b}")
print(f"Diferencia A-B (A - B):       {conjunto_a - conjunto_b}")
print(f"Diferencia B-A (B - A):       {conjunto_b - conjunto_a}")
print(f"Diferencia simétrica (A ^ B): {conjunto_a ^ conjunto_b}")
````

# 📝 Resumen: Estructuras de Datos en Python

| Estructura | Sintaxis | Ordenada | Mutable | Duplicados | Uso típico |
|-----------|---------|---------|--------|-----------|----------|
| **Lista** | `[1, 2, 3]` | ✅ | ✅ | ✅ | Colecciones generales |
| **Tupla** | `(1, 2, 3)` | ✅ | ❌ | ✅ | Datos constantes |
| **Diccionario** | `{"a": 1}` | ✅* | ✅ | ❌ (claves) | Pares clave-valor |
| **Conjunto** | `{1, 2, 3}` | ❌ | ✅ | ❌ | Valores únicos, operaciones matemáticas |

*En Python 3.7+ los diccionarios mantienen orden de inserción.

### 💡 Regla general para elegir la estructura correcta:
- ¿Necesitas modificar los datos? → **Lista o Diccionario**
- ¿Los datos no deben cambiar? → **Tupla**
- ¿Necesitas asociar claves a valores? → **Diccionario**
- ¿Necesitas elementos únicos o hacer operaciones de conjuntos? → **Conjunto**
