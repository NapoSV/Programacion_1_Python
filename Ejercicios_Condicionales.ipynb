```python
# 35. Pedir un número y determinar si es primo o no.
num = int(input("Ingrese un número: "))
es_primo = True
if num <= 1:
    es_primo = False
for i in range(2, int(num**0.5)+1):
    if num % i == 0:
        es_primo = False
        break
if es_primo:
    print("Es primo")
else:
    print("No es primo")
```

```python
# 40. Solicitar dos números y una operación (+, -, *, /) y mostrar el resultado según la opción.
a = float(input("Ingrese el primer número: "))
b = float(input("Ingrese el segundo número: "))
op = input("Ingrese la operación (+, -, *, /): ")
if op == "+":
    print("Resultado:", a + b)
elif op == "-":
    print("Resultado:", a - b)
elif op == "*":
    print("Resultado:", a * b)
elif op == "/" and b != 0:
    print("Resultado:", a / b)
else:
    print("Operación inválida")
```
